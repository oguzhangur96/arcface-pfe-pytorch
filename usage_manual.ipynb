### Preparing the Data

**1-)** Download 112x112 CASIA Webface dataset from link below:

https://drive.google.com/open?id=1Of_EVz-yHV7QVWQGihYfvtny9Ne8qXVz

**2-)** Download and edit image paths file:

https://github.com/seasonSH/Probabilistic-Face-Embeddings/blob/master/data/list_casia_mtcnncaffe_aligned_nooverlap.txt

**3-)** Download validation datasets from link below:

https://github.com/deepinsight/insightface/wiki/Dataset-Zoo

**4-)** run from terminal:
``` python prepare_val_data.py ```
to prepare validation data. More explanation could be found inside the file.


### Training
**1-)** Download pretrained MobileFaceNet model from the link below:

https://1drv.ms/u/s!AhMqVPD44cDOhkSMHodSH4rhfb5u

**2-)** Open ``` config/train_config.py ``` and edit arguments in file corresponding to your folder setup.

**3-)** Run ``` python train.py ``` to obtain trained PFE model with parameters in ``` config/train_config.py ```

### Evaluation
**1-)** You could evaluate models with ``` eval_mobilefacenet.ipynb ``` and ``` eval_pfe.ipynb ``` notebooks

### Inference
**1-)** To get inference on any face photo or do verification on two photos, with MobileFaceNet or PFE you could explore
``` eval_mobilefacenet.ipynb ```
